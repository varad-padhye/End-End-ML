```python
import os
import cv2
import numpy as np

DATASET_DIR = "../input/plantvillage-dataset/color"
IMG_SIZE = 64

images = []
labels = []
class_names = sorted(os.listdir(DATASET_DIR))

for label, class_name in enumerate(class_names):
    class_path = os.path.join(DATASET_DIR, class_name)
    
    if not os.path.isdir(class_path):
        continue
    
    for img_name in os.listdir(class_path):
        if not img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            continue
        
        img_path = os.path.join(class_path, img_name)
        img = cv2.imread(img_path)
        
        if img is None:
            continue
        
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0
        
        images.append(img)
        labels.append(label)

images = np.array(images)
labels = np.array(labels)

print("Classes:", class_names)
print("Images shape:", images.shape)
print("Labels shape:", labels.shape)
```

```python
np.save("image.npy",images)
np.save("label.npy",labels)
```

```python
path_x="/kaggle/working/image.npy"
path_y="/kaggle/working/label.npy"
X = np.load(path_x)

y=np.load(path_y)

X_flat = X.reshape(X.shape[0], -1)


```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_flat,y,test_size=0.2,random_state=42)
X_train
```

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

# 1️⃣ Create Dummy Classifier (Most Frequent strategy)
Dummy_Model = DummyClassifier(strategy="most_frequent")

# 2️⃣ Train Dummy Model
Dummy_Model.fit(X_train, y_train)

# 3️⃣ Predict on Test Set
y_pred_dummy = Dummy_Model.predict(X_test)

# 4️⃣ Print Dummy Accuracy
dummy_accuracy = accuracy_score(y_test, y_pred_dummy)
print("Dummy Accuracy:", dummy_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Step C: Create & Train Model
My_Model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

My_Model.fit(X_train, y_train)

# Predictions
y_pred = My_Model.predict(X_test)

# Step D: Evaluation
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(cm)
```
